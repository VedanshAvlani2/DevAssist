```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

model_name = "mistralai/Mistral-7B-v0.1"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16    # changed from bfloat16 to float16
)

print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"           # added — prevents subtle tensor issues

print("Loading model...")
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
    torch_dtype=torch.float16              # force fp16 at load time
)

model.config.use_cache = False             # required for gradient checkpointing
print("Model loaded.")
```

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

# Prepare model for 4-bit training
model = prepare_model_for_kbit_training(model)

# LoRA config
# r=16 means we add small 16-rank matrices alongside original weights
# Instead of training 7B params, we train ~4M params
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],  # which layers to apply LoRA to
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)

# Show how many params we are actually training
model.print_trainable_parameters()
```

```python
from datasets import load_dataset

print("Loading dataset...")
dataset = load_dataset("iamtarun/python_code_instructions_18k_alpaca", split="train")

# Use only 2000 examples — enough for good fine-tuning, trains in 30-40 min
dataset = dataset.select(range(4000))

print(f"Total examples: {len(dataset)}")

def format_example(example):
    text = f"""### Instruction:
{example['instruction']}

### Input:
{example.get('input', '')}

### Response:
{example['output']}"""
    return {"text": text}

print("Formatting dataset...")
dataset = dataset.map(format_example)
print("Done.")
```

```python
def tokenize(example):
    result = tokenizer(
        example["text"],
        truncation=True,
        max_length=512,        # max tokens per example
        padding="max_length"
    )
    result["labels"] = result["input_ids"].copy()
    return result

print("Tokenizing...")
tokenized_dataset = dataset.map(tokenize, batched=True, remove_columns=dataset.column_names)
print(f"Tokenized {len(tokenized_dataset)} examples.")
```

```python
from transformers import TrainingArguments, DataCollatorForLanguageModeling
from trl import SFTTrainer

training_args = TrainingArguments(
    output_dir="./mistral-devassist",
    num_train_epochs=1,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=8,
    warmup_steps=50,
    learning_rate=2e-4,
    fp16=False,                          # disable fp16 entirely
    bf16=False,                          # disable bf16 entirely
    logging_steps=50,
    save_steps=200,
    save_total_limit=2,
    report_to="none",
    optim="adamw_torch"                  # standard optimizer, no bf16 issues
)

trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False),
)

print("Starting training...")
trainer.train()
print("Training complete.")
```

```python
# Save only the LoRA adapter weights — much smaller than full model
model.save_pretrained("./mistral-devassist-lora")
tokenizer.save_pretrained("./mistral-devassist-lora")
print("Adapter saved to ./mistral-devassist-lora")
```
